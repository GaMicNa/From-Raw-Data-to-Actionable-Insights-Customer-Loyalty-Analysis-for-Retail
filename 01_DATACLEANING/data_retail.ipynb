**REMOVE IRRELEVANT DATA**



```python
col_drop= ["sales_commission_code", "created_at", "discount_amount", "Year", "Month", "M-Y", "BI Status", "FY", "payment_method"]
df=df.drop(columns = col_drop)
df
```

**REMOVE DUPLICATE**S

```python
df = df.drop_duplicates()
df.head(10)
```

**RENAME COLUMNS**

```python
new_names={'category_name_1' : 'category_name', 'Customer Since' : 'customer_since', 'Customer ID' : 'customer_id', 'MV' : 'total_by_item',
           'grand_total' : 'total_by_order', 'increment_id' : 'order_id', 'price' : 'unit_price', 'Working Date' : 'working_date'}
df= df.rename(columns=new_names)
df.head(10)
```

**RE ORDER COLUMNS**

```python
col_order = [ 'order_id', 'customer_id', 'item_id', 'status', 'working_date', 'unit_price', 'qty_ordered',
             'total_by_item', 'total_by_order', 'customer_since', 'category_name', 'sku' ]
df = df[col_order]
df.iloc[50020:50030]
```

**DATA TYPE INSPECTION**

```python
df.info()
```

**CHECK VALUES IN A 'ORDER_ID' COLUMN THAT DO NOT MEET THE CONDITION OF BEING NUMBERS AND CONVERT IN NULL VALUES.**

```python
def check_values(dataframe, columns):

    for colum in columns:
        dataframe[colum] = pd.to_numeric(dataframe[colum], errors='coerce')
    return dataframe

df = check_values(df, ['order_id', 'total_by_item'])

values_no_num =pd.DataFrame({'items' : df['total_by_item'], 'order' : df['order_id'] })
values_no_num = values_no_num.isnull().sum()
print(values_no_num)
```

**REMOVE NULL DATA FROM 'ORDER_ID' AND 'TOTAL_BY_ITEM', THEN VERIFY TO 'ORDER_ID' 'INT64' TYPE DATA.**

```python
df = df.dropna(axis=0, how='any')
df['order_id']= df['order_id'].astype('int64')
df.dtypes
```

**THE TOTAL PER ORDER CANNOT BE LESS THAN THE TOTAL PER ITEM, SO WE REMOVE THE ORDER_ID THAT DOES NOT MEET THE CONDITION.**

```python
df = df.loc[df['total_by_item'] <= df['total_by_order']]
df.head(10)
```

**STANDARDIZE 'WORKING_DATE' DATE FORMAT.**

```python

def standardize_date(df, column_name):

    df[column_name] = pd.to_datetime(df[column_name], errors='coerce')

    df[column_name] = df[column_name].dt.strftime('%Y-%m-%d')

    return df

df = standardize_date(df, 'working_date')
df.head(5)
```

**CONVERT 'WORKING_DATE' DATA TYPE.**

```python
df['working_date'] = df['working_date'].astype('datetime64[ns]')
df.dtypes
```

**THE 'WORKING_DATE' COLUMN CAN NOT BE EARLIER THAN 'CUSTOMER_SINCE', SO WE ARE GOING TO DIVIDE THE DATAFRAME BETWEEN DATA FORMAT AAAA-DD-MM AND AAAA-MM-DD, STANDARDIZE AND THEN JOIN TOGETHER.**

```python

df_1 = df.loc[df['working_date']<df['customer_since']]
df_2 = df.loc[df['working_date']>=df['customer_since']]

df_1['working_date'] = df_1['working_date'].dt.strftime('%Y-%d-%m')
df_1['working_date'] = df_1['working_date'].astype('datetime64')

df = pd.concat([df_1, df_2])
#df_1
```

**SORT COLUMNS BY "ORDER_ID" VALUES.**

```python
df = df.sort_values(by='order_id')
df = df.reset_index(drop = True)
df.tail(10)
```

**FINDING UNIQUE VALUES.**

```python
def unique_values(df, columns):
    v_unique = {}
    for column in columns:
        v_unique[column] = df[column].nunique()
    return v_unique

columns_data = ['item_id', 'order_id', 'customer_id']
result = unique_values(df, columns_data)

for columns, unique_values in result.items():
    print(f"Unique values by column '{columns}': {unique_values}")
```
